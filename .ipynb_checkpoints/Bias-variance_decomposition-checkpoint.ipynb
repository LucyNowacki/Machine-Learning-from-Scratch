## 1. Polynomial Regression in PyTorch
Create your own dataset by defining function $f(x):=sin(2x)$, and by sampling the function at points $\{x_i\}^s_{i=1}$ from the interval [0,1], computing corresponding output samples $y_{i}=\hat{f}(x_i) \,\,\,\forall \,\, i\in\{1,...,s\}$. Perturb your data with the normal distribution random noise with mean zero and variance $\sigma^2$ to obtain noisy simulated data samples $\{y^\epsilon_i\}^s_{i=1}$.

```python
plot_fontsize = 11
no_of_samples = 50
noise_standard_deviation = 0.1
input_data = torch.linspace(0, 1, steps = no_of_samples, dtype=torch.float64)
true_output_data = torch.sin(2*np.pi*input_data)
torch.manual_seed(13)
output_data = true_output_data + noise_standard_deviation * torch.randn(no_of_samples, dtype=torch.float64)
ax = plt.subplot()
ax.plot(input_data, true_output_data, label='Ground truth') 
ax.plot(input_data, output_data, '*', label='Data samples')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
ax.legend(loc='upper left', fontsize=plot_fontsize)
plt.show()
```

In the following, we define the function **polynomially_augmented_data_matrix** that sets up the data matrix , given arguments *input_data* and *polynomial_degree*. We also define functions **polynomial_model_function** and **least_squares_solution** that evaluate $\Phi(X) w$, where $\Phi(X)$ denotes the augmented data matrix and $w$ some weights, and solve $\Phi(X)^\top \Phi(X) w = \Phi(X)^\top y$ respectively.

```python
def polynomially_augmented_data_matrix(input_data, polynomial_degree=1):
    data_matrix = input_data.view(-1, 1).pow(0)
    for j in range(polynomial_degree):
        data_matrix = torch.cat((data_matrix, input_data.view(-1, 1).pow(j + 1)), 1)
    return data_matrix

def polynomial_model_function(augmented_data_matrix, weights):
    return torch.mv(augmented_data_matrix, weights)

def least_squares_solution(augmented_data_matrix, output_data):
    gram_matrix = torch.mm(augmented_data_matrix.t(), augmented_data_matrix)
    transformed_output_data = torch.mv(augmented_data_matrix.t(), output_data)
    optimal_weights, _ = torch.solve(transformed_output_data.view(-1, 1), gram_matrix)
    return optimal_weights.view(-1)
```

We test the functions for a polynomial of degree 21 with the following code snippet

```python
polynomial_degree = 21
augmented_data_matrix = polynomially_augmented_data_matrix(input_data, polynomial_degree)
optimal_weights = least_squares_solution(augmented_data_matrix, output_data)
model_output = polynomial_model_function(augmented_data_matrix, optimal_weights)
```

and visualise the model approximation together with the true function and the noisy data samples:

```python
ax = plt.subplot()
ax.plot(input_data, true_output_data, label='Ground truth')
ax.plot(input_data, output_data, '*', label='Data samples')
ax.plot(input_data, model_output, '--', label='Model fit')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
ax.legend(loc='lower left', fontsize=plot_fontsize)
plt.show()
```

## 2. Bias-variance decomposition of a polynomial ridge regression


```python
plot_fontsize = 11
no_of_sets = 100
no_of_samples_per_set = 200
no_of_total_samples = no_of_samples_per_set * no_of_sets
noise_standard_deviation = 0.5
torch.manual_seed(13)
training_input_data = torch.linspace(0, 1, steps = no_of_total_samples, dtype=torch.float64)
training_input_data = training_input_data[torch.randperm(no_of_total_samples)]
training_input_data, _ = torch.sort(training_input_data.view(no_of_sets, no_of_samples_per_set), 1)
training_output_data = torch.sin(2*np.pi*training_input_data) + noise_standard_deviation * \
        torch.randn(no_of_sets, no_of_samples_per_set, dtype=torch.float64)
```

We define a finely discretised validation dataset and visualise both the training and the validation data.

```python
no_of_validation_samples = 800
validation_input_data = torch.linspace(0, 1, steps = no_of_validation_samples, dtype=torch.float64)
true_output_data = torch.sin(2*np.pi*validation_input_data)

fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()
ax = plt.subplot()
ax.plot(training_input_data, training_output_data, '-', color='orange', linewidth=0.1, label='Validation data')
ax.plot(validation_input_data, true_output_data, color='blue', label='Ground truth')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
plt.show()
```

Let's write the function computing the solution of 
$$w_t=arg \underset{w \in \mathbb{R}^{d+1}}{\min}\left\{\frac{1}{2}\|\Phi(X)w-y\|^2+\frac{\gamma}{2}\|w\|^2   \right\}$$
or equivalently the linear system
$$\left(\Phi(X)^T\Phi(X)+\gamma I\right)\,w_t= \Phi(X)^T y$$
$w_t$ is the mathematical notation of the return variable *weights*.
<br>
The argument *gram_matrix* corresponds to the matrix $\Phi(X)^T\Phi(X) $, wheres *transformed_output_data* refers to $\Phi(X)^T y$ and the regualrisation parameter is $\gamma$.<br>
First, we define a whole range of noisy data samples from $\sin(2\pi x)$. We specifiy a fixed number of sets *no_of_sets* where each set contains *no_of_samples_per_set* samples, stored in the variables *training_input_data* and *training_output_data*.

```python
def ridge_regression_least_squares(gram_matrix, transformed_output_data, regularisation_parameter):
    shifted_gram_matrix = gram_matrix + regularisation_parameter * torch.eye(gram_matrix.size(0))
    weights, _ = torch.solve(transformed_output_data.view(-1, 1), shifted_gram_matrix)
    return weights.view(-1)
```

With the followin code snippet, we train optimal weights for the different data samples.

```python
polynomial_degree = 31
regularisation_parameter_grid_size = 200
regularisation_parameters = torch.logspace(-7, 1, steps=regularisation_parameter_grid_size)

weights = torch.zeros(polynomial_degree + 1, no_of_sets, regularisation_parameter_grid_size, dtype=torch.float64)
for counter2 in range(regularisation_parameter_grid_size):
    for counter1 in range(no_of_sets):
        training_data_matrix = polynomially_augmented_data_matrix(training_input_data[counter1, :], polynomial_degree)
        gram_matrix = torch.mm(training_data_matrix.t(), training_data_matrix)
        transformed_output_data = torch.mv(training_data_matrix.t(), training_output_data[counter1, :])
        weights[:, counter1, counter2] = ridge_regression_least_squares(gram_matrix, transformed_output_data, \
                                                                        regularisation_parameters[counter2])
```

We can validate the trained parameters by applying the trained models to the validation data, and by visualising different model approximations.

```python
validation_matrix = polynomially_augmented_data_matrix(validation_input_data, polynomial_degree)
model_outputs = torch.mm(validation_matrix, weights.view(polynomial_degree + 1, -1)).view(-1, no_of_sets, \
                                        regularisation_parameter_grid_size)
expected_model_output = model_outputs.mean(1)
print(model_outputs.size(), expected_model_output.size())

fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()
ax.plot(validation_input_data.view(no_of_validation_samples, 1).expand(no_of_validation_samples, no_of_samples_per_set), \
         expected_model_output, color='red', linewidth=0.1)
ax.plot(validation_input_data, true_output_data, color='blue', label='Ground truth')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
plt.show()
```

With the following code, we can visualise the average squared bias

$$ \text{Bias}_t(f_{w_t}(x))^2 = \left(\mathbb{E}_t[f_{w_t}(x)] - \hat f(x) \right)^2 $$

for all regularisation parameters.

```python
bias = expected_model_output - true_output_data.view(no_of_validation_samples, 1).expand(no_of_validation_samples, \
                                                regularisation_parameter_grid_size)
plt.plot(regularisation_parameters, bias.pow(2).mean(0))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('regularisation parameter', fontsize=plot_fontsize)
plt.ylabel('average of squared bias', fontsize=plot_fontsize)
#plt.savefig('Images/squared-bias.pdf')
```

And we do the same for the average variance

$$ \text{Var}_t[f_{w_t}(x)] = \mathbb{E}_t\left[ \left( f_{w_t}(x) - \mathbb{E}_t\left[f_{w_t}(x)\right] \right)^2 \right]$$

for all regularisation parameters.

```python
variance = (model_outputs - expected_model_output.view(no_of_validation_samples, 1, \
                    regularisation_parameter_grid_size).expand(model_outputs.size())).pow(2).mean(1)
plt.plot(regularisation_parameters, variance.mean(0))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('regularisation parameter', fontsize=plot_fontsize)
plt.ylabel('average of variance', fontsize=plot_fontsize)
#plt.savefig('Images/variance.pdf')
```

We then sum the averaged squared bias, variance and the noise variance to plot the average mean squared error.

```python
average_mean_squared_error = bias.pow(2).mean(0).numpy() + variance.mean(0).numpy() + noise_standard_deviation ** 2
print(regularisation_parameters[average_mean_squared_error.argmin()])
plt.plot(regularisation_parameters, average_mean_squared_error)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('regularisation parameter', fontsize=plot_fontsize)
plt.ylabel('averaged mean squared error', fontsize=plot_fontsize)
#plt.savefig('Images/bias-variance-decomposition-without-noise-variance.pdf')
```

We conclude this part by plotting all quantities in one plot.

```python
fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()
ax.plot(regularisation_parameters, bias.pow(2).mean(0), label='average squared bias')
ax.plot(regularisation_parameters, variance.mean(0), label='average variance')
ax.plot(regularisation_parameters, (noise_standard_deviation ** 2) * torch.ones(regularisation_parameter_grid_size), \
         label='noise variance')
ax.plot(regularisation_parameters, average_mean_squared_error, '--', label='average mean squared error')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('regularisation parameter', fontsize=plot_fontsize)
ax.legend(loc='best', fontsize=plot_fontsize)
#plt.savefig('Images/bias-variance-decomposition.pdf')
plt.show()
```

## 3. Optimal regularisation parameter selection for the validation error

Here,we define one additional function **mean_squared_error** and the class **ridge_regression**. The class **ridge_regression** is a sub-class of the neural network base-class **nn.Module**, which usually consists of a constructor and a routine named *forward* that evaluates the network. In our case it should simply call the routine **ridge_regression_least_squares**.

```python
def mean_squared_error(argument1, argument2):
    return 1/2 * torch.mean( ( argument1 - argument2 ) ** 2)

class ridge_regression(nn.Module):
    def __init__(self, gram_matrix, transformed_output_data):
        super(ridge_regression, self).__init__()
        self.gram_matrix = gram_matrix
        self.transformed_output_data = transformed_output_data
        self.regularisation_parameter = nn.Parameter((10 ** (-1)) * torch.ones(1, dtype=torch.float64))
        
    def forward(self):
        return ridge_regression_least_squares(self.gram_matrix, self.transformed_output_data, self.regularisation_parameter)
```

## Initialise data

We initialise some training data for the ridge regression model and some validation data in order to optimise for the regularisation parameter. Subsequently, we visualise  those data points.

```python
plot_fontsize = 11
no_of_training_samples = 50
noise_standard_deviation = 0.5
training_input_data = torch.linspace(0, 1, steps = no_of_training_samples, dtype=torch.float64)
true_output_data = torch.sin(2*np.pi*training_input_data)
torch.manual_seed(13)
training_output_data = true_output_data + noise_standard_deviation * torch.randn(no_of_training_samples, \
                                                                                 dtype=torch.float64)
no_of_validation_samples = 1000
validation_input_data = torch.linspace(0, 1, steps = no_of_validation_samples, dtype=torch.float64)
torch.manual_seed(14)
validation_output_data = torch.sin(2*np.pi*validation_input_data) + noise_standard_deviation * \
                            torch.randn(no_of_validation_samples, dtype=torch.float64)

fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()
ax.plot(training_input_data, training_output_data, '*', label='Training samples')
ax.plot(validation_input_data, validation_output_data, '*', label='Validation samples')
ax.plot(training_input_data, true_output_data, label='Ground truth')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
ax.legend(loc='best', fontsize=plot_fontsize)
```

## Initialise model

We now initialise our ridge regression model with a polynmial of degree 31.

```python
polynomial_degree = 31
training_data_matrix = polynomially_augmented_data_matrix(training_input_data, polynomial_degree)
validation_data_matrix = polynomially_augmented_data_matrix(validation_input_data, polynomial_degree)
gram_matrix = torch.mm(training_data_matrix.t(), training_data_matrix)
transformed_output_data = torch.mv(training_data_matrix.t(), training_output_data)
regularisation = ridge_regression(gram_matrix, transformed_output_data)
```

As a sanity check we apply the initialisation of our model and **ridge_regression_least_squares** with regularisation parameter $\alpha = 10^{-1}$ in order to see that they coincide.

```python
weights1 = regularisation()
weights2 = ridge_regression_least_squares(gram_matrix, transformed_output_data, (10 ** (-1)))
model_output1 = polynomial_model_function(validation_data_matrix, weights1)
model_output2 = polynomial_model_function(validation_data_matrix, weights2)

fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()
ax.plot(training_input_data, true_output_data, label='Ground truth')
ax.plot(validation_input_data, model_output1.detach(), '--', label='Model output 1')
ax.plot(validation_input_data, model_output2.detach(), '*', label='Model output 2')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
ax.legend(loc='best', fontsize=plot_fontsize)
plt.show()
```

#### Now we compute our regularisation parameter via gradient descent.

```python
outer_step_size = 10 ** (-4)
maximum_no_of_iterations = 10000

optimiser = optim.SGD(regularisation.parameters(), lr=outer_step_size)

losses = []
for outer_counter in range(maximum_no_of_iterations):
    optimiser.zero_grad()
    weights = regularisation()
    model_output = polynomial_model_function(validation_data_matrix, weights)
    validation_error = mean_squared_error(model_output, validation_output_data)
    losses.append(validation_error.item())
    validation_error.backward()
    optimiser.step()    
    if outer_counter % 1000 == 0:
        print("Gradient descent({it}/{max_iter}): loss = {l}".format(
                  it=outer_counter, max_iter=maximum_no_of_iterations - 1, l=losses[outer_counter]))
losses.append(validation_error.item())        

print("Gradient descent completed: regularisation parameter = {alpha}, loss = {l}.".format(
    alpha = regularisation.regularisation_parameter.detach().numpy(), l=validation_error.item()))
```

We conclude by visualising the model ouput with the estimated regularisation parameter.

```python
fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot()

ax.plot(training_input_data, true_output_data, label='Ground truth')
ax.plot(validation_input_data, validation_output_data, '*', label='Validation samples')
ax.plot(validation_input_data, model_output.detach(), '--', label='Model output')
ax.set_xlabel('x-axis', fontsize=plot_fontsize)
ax.set_ylabel('y-axis', fontsize=plot_fontsize)
ax.legend(loc='best', fontsize=plot_fontsize)
plt.show()
```

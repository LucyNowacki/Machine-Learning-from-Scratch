```python
class AddGaussianNoise(object):
    def __init__(self, mean=0., std=1.):
        self.std = std
        self.mean = mean
        
    def __call__(self, tensor):
        return tensor, (tensor + torch.randn(tensor.size()) * self.std + self.mean)
    
    def __repr__(self):
        return self.__class__.__name__ + '(mean={0}, std={1})'.format(self.mean, self.std)
```

### Linear autoencoder

```python
full_batch = next(iter(full_loader))[0].to(device)
```

```python
class linear_autoencoder():
    
    def __init__(self, training_data, dimension):
        decoder, _, _ = torch.svd(training_data)
        self.decoder_matrix = decoder[:, 0:dimension]
        
    def encoder(self, input):
        return self.decoder_matrix.T @ input
    
    def decoder(self, input):
        return self.decoder_matrix @ input
    
    def evaluate(self, input):
        return self.decoder(self.encoder(input))
```

```python
autoencoder_1 = linear_autoencoder(full_batch.view(60000, 784).T, 392)
```

```python
codes = autoencoder_1.encoder(full_batch.view(60000, 784).T)
print(codes.shape)
approximations = autoencoder_1.decoder(codes).T.view(60000, 1, 28, 28)
plt.subplot(1, 2, 1)
plt.imshow(full_batch[3][0].detach().cpu().numpy())
plt.subplot(1, 2, 2)
plt.imshow(approximations[3][0].detach().cpu().numpy())
```

### Neural network autoencoder

```python
class autoencoder(nn.Module):
    def __init__(self):
        super(autoencoder, self).__init__()
        self.encoder_convolution_1 = nn.Parameter(0.1 * torch.randn(10, 1, 5, 5))
        self.encoder_convolution_2 = nn.Parameter(0.1 * torch.randn(3, 10, 3, 3))
        self.encoder_bias_1 = nn.Parameter(0.1 * torch.randn(10))
        self.encoder_bias_2 = nn.Parameter(0.1 * torch.randn(3))
        self.encoder_transform = nn.Parameter(0.1 * torch.randn(588, 392))
        #self.encoder_activation = nn.PReLU(init=0.05)        
        #self.encoder_activation = nn.ReLU()
        #self.encoder_activation = nn.Softshrink(lambd=0.5)
        self.encoder_activation = nn.Tanh()        
        self.encoder_batch_norm_1 = nn.BatchNorm2d(10)
        self.encoder_batch_norm_2 = nn.BatchNorm2d(3)
        self.encoder_batch_norm_3 = nn.BatchNorm1d(392)
        self.decoder_batch_norm_1 = nn.BatchNorm1d(588)
        self.decoder_batch_norm_2 = nn.BatchNorm2d(10)
        self.decoder_batch_norm_3 = nn.BatchNorm2d(1)
        #self.decoder_activation = nn.PReLU(init=0.05)
        #self.decoder_activation = nn.ReLU()
        #self.decoder_activation = nn.Softshrink(lambd=0.5)
        self.decoder_activation = nn.Tanh()        
        self.decoder_transform = nn.Parameter(0.1 * torch.randn(392, 588))
        self.decoder_convolution_1 = nn.Parameter(0.1 * torch.randn(3, 10, 3, 3))
        self.decoder_convolution_2 = nn.Parameter(0.1 * torch.randn(10, 1, 5, 5))
        self.decoder_bias_1 = nn.Parameter(0.1 * torch.randn(10))
        self.decoder_bias_2 = nn.Parameter(0.1 * torch.randn(1))
    
    def encoder(self, x):
        z = fct.conv2d(x, self.encoder_convolution_1, self.encoder_bias_1, padding=2)
        z = self.encoder_batch_norm_1(z)
        z = self.encoder_activation(z)
        z = fct.interpolate(z, [14, 14], mode='bilinear')
        z = fct.conv2d(z, self.encoder_convolution_2, self.encoder_bias_2, padding=1)
        z = self.encoder_batch_norm_2(z)
        z = self.encoder_activation(z)
        z = torch.matmul(z.view(-1, 588), self.encoder_transform)
        z = self.encoder_batch_norm_3(z)
        z = self.encoder_activation(z)
        return z
    
    def decoder(self, z):
        x = torch.matmul(z, self.decoder_transform)
        x = self.decoder_batch_norm_1(x)
        x = self.decoder_activation(x.view(-1, 3, 14, 14))
        x = fct.conv_transpose2d(x, self.decoder_convolution_1, self.decoder_bias_1, padding=1)
        x = fct.interpolate(x, [28, 28], mode='bilinear')
        x = self.decoder_batch_norm_2(x)
        x = self.decoder_activation(x)
        x = fct.conv_transpose2d(x, self.decoder_convolution_2, self.decoder_bias_2, padding=2)
        x = self.decoder_batch_norm_3(x)
        x = self.encoder_activation(x)
        return x
    
    def forward(self, x):
        return self.decoder(self.encoder(x))
```

```python
network = autoencoder().to(device)
```

```python
def train_network_parameters(network, loader):
    
    criterion = nn.MSELoss()
    number_of_epochs = 50
    optimiser = optim.Adam(network.parameters())

    running_loss = []

    for counter in range(number_of_epochs):
    
        start = time.time()
    
        epoch_loss = 0
    
        for inner_counter, (inputs, _) in enumerate(loader):
            inputs = inputs.to(device)
            #outputs = outputs.to(device)
        
            optimiser.zero_grad()        
            outputs = network(inputs)
            loss = criterion(inputs, outputs)
            loss.backward()
            optimiser.step() 
            epoch_loss += loss.item()
        
        running_loss.append(1/len(loader) * epoch_loss)
    
        end = time.time()
        
        if (counter + 1) % 1 == 0: 
            print('Epoch [%d/%d], Loss: %.4f, Time: %.2f s' 
                   %(counter + 1, number_of_epochs, loss.item(), end - start))

    print('Epoch [%d/%d] completed, Loss: %.4f' 
                   %(counter + 1, number_of_epochs, loss.item()))

    network.eval()
    
    return network, running_loss
```

```python
network, running_loss = train_network_parameters(network, train_loader)
```

### Autoencoder with weight-sharing

```python
class weight_sharing_autoencoder(nn.Module):
    def __init__(self):
        super(weight_sharing_autoencoder, self).__init__()
        self.convolution_1 = nn.Parameter(0.1 * torch.randn(10, 1, 5, 5))
        self.convolution_2 = nn.Parameter(0.1 * torch.randn(3, 10, 3, 3))
        self.transform = nn.Parameter(0.1 * torch.randn(588, 392))
        self.encoder_bias_1 = nn.Parameter(0.1 * torch.randn(10))
        self.encoder_bias_2 = nn.Parameter(0.1 * torch.randn(3))        
        #self.encoder_activation = nn.PReLU(init=0.05)        
        #self.encoder_activation = nn.ReLU()
        #self.encoder_activation = nn.Softshrink(lambd=0.5)
        self.encoder_activation = nn.Tanh()        
        self.encoder_batch_norm_1 = nn.BatchNorm2d(10)
        self.encoder_batch_norm_2 = nn.BatchNorm2d(3)
        self.encoder_batch_norm_3 = nn.BatchNorm1d(392)
        self.decoder_batch_norm_1 = nn.BatchNorm1d(588)
        self.decoder_batch_norm_2 = nn.BatchNorm2d(10)
        self.decoder_batch_norm_3 = nn.BatchNorm2d(1)
        #self.decoder_activation = nn.PReLU(init=0.05)
        #self.decoder_activation = nn.ReLU()
        #self.decoder_activation = nn.Softshrink(lambd=0.5)
        self.decoder_activation = nn.Tanh()        
        self.decoder_bias_1 = nn.Parameter(0.1 * torch.randn(10))
        self.decoder_bias_2 = nn.Parameter(0.1 * torch.randn(1))
    
    def encoder(self, x):
        z = fct.conv2d(x, self.convolution_1, self.encoder_bias_1, padding=2)
        z = self.encoder_batch_norm_1(z)
        z = self.encoder_activation(z)
        z = fct.interpolate(z, [14, 14], mode='bilinear')
        z = fct.conv2d(z, self.convolution_2, self.encoder_bias_2, padding=1)
        z = self.encoder_batch_norm_2(z)
        z = self.encoder_activation(z)
        z = torch.matmul(z.view(-1, 588), self.transform)
        z = self.encoder_batch_norm_3(z)
        z = self.encoder_activation(z)
        return z
    
    def decoder(self, z):
        x = torch.matmul(z, self.transform.T)
        x = self.decoder_batch_norm_1(x)
        x = self.decoder_activation(x.view(-1, 3, 14, 14))
        x = fct.conv_transpose2d(x, self.convolution_2, self.decoder_bias_1, padding=1)
        x = fct.interpolate(x, [28, 28], mode='bilinear')
        x = self.decoder_batch_norm_2(x)
        x = self.decoder_activation(x)
        x = fct.conv_transpose2d(x, self.convolution_1, self.decoder_bias_2, padding=2)
        x = self.decoder_batch_norm_3(x)
        x = self.encoder_activation(x)
        return x
    
    def forward(self, x):
        return self.decoder(self.encoder(x))
```

```python
weight_sharing_network = weight_sharing_autoencoder().to(device)
weight_sharing_network, weight_sharing_running_loss = train_network_parameters(weight_sharing_network, \
                                                                               train_loader)
```
